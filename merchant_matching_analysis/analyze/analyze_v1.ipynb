# Introduction

```python
import pandas
import numpy
import matplotlib.pyplot as plt
import seaborn
```

```python
df = pandas.read_csv('./result_v1.csv')
```

# Group by score and count the number plot in bar chart

```python
# Group by similarity_score and count the number and plot it using bar chart
df.groupby('similarity_score').count()['qr_marchant_name'].plot.bar()
plt.show()
```

# Group by score and count the number plot in pie chart

```python
# Group by similarity_score and count the number and plot it using pie chart
df.groupby('similarity_score').count()['qr_marchant_name'].plot.pie(autopct='%.0f%%')
plt.show()
```

# The count of matched merchants name

```python
# Count the number of similarity_score which is match
df[df['similarity_score'] == 1].count()['qr_marchant_name']
```

# Group by name and count which name is most popular in bar chart

```python
# Group by qr_marchant_name and count which name is most frequent using bar chart
df.groupby('qr_marchant_name').count()['similarity_score'].sort_values(ascending=False).head(10).plot.bar()
plt.show()
```

# Group by name and count which name is most popular in pie chart

```python
# Group by qr_marchant_name and count which name is most frequent using pie chart
df.groupby('qr_marchant_name').count()['similarity_score'].sort_values(ascending=False).head(10).plot.pie(autopct='%.0f%%')
plt.show()
```

# Name that got the highest amount of testing

```python
# what qr_marchant_name get the highest amount of testing
df.groupby('qr_marchant_name').count()['similarity_score'].sort_values(ascending=False).head(10)
```
